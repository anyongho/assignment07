## import packages for plotting graphs and manipulate data
### And write matplotlib incline for plot error

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

num     = 1001
std     = 5 

# x  : x-coordinate data
# y1 : (noisy) y-coordinate data
# y2 : (clean) y-coordinate data 

def fun(x):
 
	# f = np.sin(x) * (1 / (1 + np.exp(-x))) 
	f = np.abs(x) * np.sin(x)

	return f


n       = np.random.rand(num)
nn      = n - np.mean(n)
x       = np.linspace(-10,10,num)
y1      = fun(x)
y2      = y1 + nn * std

plt.plot(x, y1, 'b.', x, y2, 'k.')
plt.show()
```

## This is for polynomial 0~9 plot

```python
fp1 = np.polyfit(x, y1, 0)
fp2 = np.polyfit(x,y1,1)
fp3 = np.polyfit(x,y1,2)
fp4 = np.polyfit(x,y1,3)
fp5 = np.polyfit(x,y1,4)
fp6 = np.polyfit(x,y1,5)
fp7 = np.polyfit(x,y1,6)
fp8 = np.polyfit(x,y1,7)
fp9 = np.polyfit(x,y1,8)
fp10 = np.polyfit(x,y1,9)

f1 = np.poly1d(fp1)
f2 = np.poly1d(fp2)
f3 = np.poly1d(fp3)
f4 = np.poly1d(fp4)
f5 = np.poly1d(fp5)
f6 = np.poly1d(fp6)
f7 = np.poly1d(fp7)
f8 = np.poly1d(fp8)
f9 = np.poly1d(fp9)
f10 = np.poly1d(fp10)

x       = np.linspace(-10,10,num)
plt.plot(x, f1(x), lw=2, color = 'r', label = '0 polyfit')
plt.plot(x, f2(x), lw=2, color = 'b', label = '1 polyfit')
plt.plot(x, f3(x), lw=2, color = 'y', label = '2 polyfit')
plt.plot(x, f4(x), lw=2, color = 'g', label = '3 polyfit')
plt.plot(x, f5(x), lw=2, color = 'c', label = '4 polyfit')
plt.plot(x, f6(x), lw=2, color = 'm', label = '5 polyfit')
plt.plot(x, f7(x), lw=2, color = 'k', label = '6 polyfit')
plt.plot(x, f8(x), lw=2, color = 'lime', label = '7 polyfit')
plt.plot(x, f9(x), lw=2, color = 'maroon', label = '8 polyfit')
plt.plot(x, f10(x), lw=2, color = 'orange', label = '9 polyfit')
plt.legend()
plt.show()
```

## This is for least fit(least error) = polynomial 9

```python
plt.plot(x, f10(x), lw=4, color = 'red', label = '9 polyfit')
plt.plot(x, y2, 'k.')
plt.legend()
plt.show()
```
